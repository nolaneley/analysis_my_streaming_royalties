```python
import pandas as pd

df = pd.read_csv('anonymized_royalties.csv')
print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Change Quantity column to int type
df.Quantity = df.Quantity.astype(int)

# Remove rows with negative values
df = df[df.Quantity > 0]
df = df[df.Price > 0]
df = df[df.Subtotal > 0]

print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Remove downloads from data
df = df[df['Transaction Type'] != 'Download']

print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Clean up partner names
df.loc[df['Partner Name'].str.contains('Amazon'), 'Partner Name'] = 'Amazon'
df.loc[df['Partner Name'].str.contains('Apple'), 'Partner Name'] = 'Apple'
df.loc[df['Partner Name'].str.contains('Pandora'), 'Partner Name'] = 'Pandora'
df.loc[df['Partner Name'].str.contains('Google'), 'Partner Name'] = 'Google'
df.loc[df['Partner Name'].str.contains('iTunes'), 'Partner Name'] = 'iTunes'
df.loc[df['Partner Name'].str.contains('Youtube'), 'Partner Name'] = 'Youtube'
df.loc[df['Partner Name'].str.contains('YouTube'), 'Partner Name'] = 'Youtube'

df['Partner Name'].unique()
```

```python
# Save cleaned data
df.to_csv('royalties_cleaned.csv', index=False)
```

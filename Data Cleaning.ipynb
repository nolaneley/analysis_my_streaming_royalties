```python
import pandas as pd

df = pd.read_csv('anonymized_royalties.csv')
df['Sales Date'] = df['Sales Date'].astype('datetime64[ns]')
print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Change Quantity column to int type
df.Quantity = df.Quantity.astype(int)

# Remove rows with negative values
df = df[df.Quantity > 0]
df = df[df.Price > 0]
df = df[df.Subtotal > 0]

print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Remove downloads from data
df = df[df['Transaction Type'] != 'Download']

print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Clean up partner names
df.loc[df['Partner Name'].str.contains('Amazon'), 'Partner Name'] = 'Amazon'
df.loc[df['Partner Name'].str.contains('Apple'), 'Partner Name'] = 'Apple'
df.loc[df['Partner Name'].str.contains('Pandora'), 'Partner Name'] = 'Pandora'
df.loc[df['Partner Name'].str.contains('Google'), 'Partner Name'] = 'Google'
df.loc[df['Partner Name'].str.contains('iTunes'), 'Partner Name'] = 'iTunes'
df.loc[df['Partner Name'].str.contains('Youtube'), 'Partner Name'] = 'Youtube'
df.loc[df['Partner Name'].str.contains('YouTube'), 'Partner Name'] = 'Youtube'

df['Partner Name'].unique()
```

```python
# Check countries
df['Delivery Country'].unique()

countries_cleaned = ['US', 'KR', 'CA', 'JP', 'DE', 'GB', 'ES', 'BR', 'MX', 'NL', 'AU',
       'EC', 'CL', 'PH', 'FI', 'NO', 'FR', 'SE', 'MY', 'ID', 'NZ', 'AR',
       'PE', 'IT', 'UA', 'CH', 'IL', 'CZ', 'HK', 'RU', 'TW', 'TH', 'PL',
       'SG', 'GT', 'DK', 'CO', 'GR', 'HU', 'IE', 'SK', 'BE', 'PT', 'TR',
       'LT', 'LV', 'AT', 'LU', 'EE', 'DO', 'PY', 'CR', 'HR', 'BG', 'SV',
       'ZA', 'IN', 'SA', 'GE', 'UK', 'VN', 'EG', 'RO', 'BY', 'SI', 'IS',
       'MA', 'PA', 'MT', 'CY', 'PK', 'NI', 'BH', 'BO', 'TN', 'UY', 'AE',
       'HN', 'NG', 'FJ', 'QA', 'RS', 'KW', 'PR', 'GU', 'BA', 'KZ', 'MK', 'VE',
       'JO', 'BM', 'MD', 'CN', 'MU', 'BD', 'AZ', 'KE', 'DZ', 'AL', 'IQ',
       'LB', 'LK', 'BN', 'MO', 'MN', 'AM', 'KH', 'TT', 'JM', 'VI', 'ME',
       'OM', 'KG', 'GH', 'NP', 'XK', 'SL', 'LA', 'MZ', 'BS', 'AD']

df = df[df['Delivery Country'].isin(countries_cleaned)]

print("Number of rows: ", df.shape[0])
df.round(5).head()
```

```python
# Remove report date column
df.drop('Report Date', axis='columns', inplace=True)

df.round(5).head()
```

```python
# Save cleaned data
df.to_csv('royalties_cleaned.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('20230721-DDSales.txt', sep='\t')
df['Sales Date'] = df['Sales Date'].astype('datetime64[ns]')
df.head()
```

```python
df.drop(['Isrc','Barcode','CDBabySku','Artist Name'], axis='columns', inplace=True)
df.head()
```

```python
albums = df['Album Name'].unique()
for i, album in enumerate(albums):
    cur_df = df[df['Album Name'] == album]
    tracks = cur_df['Track Name'].unique()
    df.loc[df['Album Name'] == album, 'Album Name'] = 'Album_' + str(i+1)
    for j, track in enumerate(tracks):
        df.loc[df['Track Name'] == track, 'Track Name'] = 'Track_' + str(i+1) + '_' + str(j+1)
        
df.head()
```

```python
# Save data
df.to_csv('anonymized_royalties.csv', index=False)
```

```python
import pandas as pd

df = pd.read_csv('20230721-DDSales.txt', sep='\t', parse_dates=['Sales Date', 'Report Date'])
df.head()
```

```python
df.drop(['Isrc','Barcode','CDBabySku'], axis='columns', inplace=True)
df.Quantity = df.Quantity.astype(int)
df.head()
```

```python
df['Artist Name'] = 'Band Name'
df.head()
```

```python
albums = df['Album Name'].unique()
for i, album in enumerate(albums):
    cur_df = df[df['Album Name'] == album]
    tracks = cur_df['Track Name'].unique()
    df.loc[df['Album Name'] == album, 'Album Name'] = 'Album_' + str(i+1)
    for j, track in enumerate(tracks):
        df.loc[df['Track Name'] == track, 'Track Name'] = 'Track_' + str(i+1) + '_' + str(j+1)
        
df.head()
```

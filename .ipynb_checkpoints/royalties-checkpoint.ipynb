```python
import pandas as pd
import numpy as np

df = pd.read_csv('20230721-DDSales.txt', sep='\t', parse_dates=['Sales Date'])
df.round(5).head()
```

```python
df.drop(['Isrc','Barcode','CDBabySku'], axis='columns', inplace=True)
df.Quantity = df.Quantity.astype(int)
df.head()
```

```python
# Total number of all plays
sum(df.Quantity.values)
```

```python
# Total payout
sum(df.Subtotal.values)
```

```python
print("Unique countries: ")
countries = df['Delivery Country'].unique()
print(sorted(countries))
```

```python
print(df['Transaction Type'].unique())
```

```python
partners = df['Partner Name'].unique()
print(sorted(partners))
```

```python
for partner in sorted(partners):
    num_streams = sum(df[df['Partner Name'] == partner].Quantity)
    print('Partner:', partner, '-', num_streams)
```

```python
# Look at streaming only (no downloads)
df = df[df['Transaction Type']=='Stream']

# Remove any negative prices
df = df[df['Price']>0]

# Unify apple partner names
cond = df['Partner Name'].str.contains('Apple')
df.loc[cond, 'Partner Name'] = 'Apple'

# Unify amazon partner names
# cond = df['Partner Name'].str.contains('Amazon')
# df.loc[cond, 'Partner Name'] = 'Amazon'

# Get apple and spotify as separate dataframes for easier processing later
apple = df[df['Partner Name']=='Apple']
spotify = df[df['Partner Name']=='Spotify']
# amazon = df[df['Partner Name']=='Amazon']
```

```python
# Clean up partners
df.loc[df['Partner Name'].str.contains('Amazon'), 'Partner Name'] = 'Amazon'
df.loc[df['Partner Name'].str.contains('Apple'), 'Partner Name'] = 'Apple'
df.loc[df['Partner Name'].str.contains('Pandora'), 'Partner Name'] = 'Pandora'
df.loc[df['Partner Name'].str.contains('Google'), 'Partner Name'] = 'Google'
df.loc[df['Partner Name'].str.contains('iTunes'), 'Partner Name'] = 'iTunes'
df.loc[df['Partner Name'].str.contains('Youtube'), 'Partner Name'] = 'Youtube'
df.loc[df['Partner Name'].str.contains('YouTube'), 'Partner Name'] = 'Youtube'

df.head()
```

```python
# Pie chart
partners = df['Partner Name'].unique()
d = {}
plays = []
for partner in partners:
    plays.append(sum(df[df['Partner Name'] == partner].Quantity))
    
plays.sort(reverse=True)
thresh = plays[4]

for partner in partners:
    cur_plays = sum(df[df['Partner Name'] == partner].Quantity)
    if cur_plays >= plays[4]:
        d[partner] = cur_plays

rest = sum(plays[5:])
d['Other'] = rest

plt.pie(d.values(), labels=d.keys(), autopct='%1.1f%%',
       pctdistance=0.9, labeldistance=1.25)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Histogram and Spotify and Apple payments
sns.displot(apple.Price, kde=True)
plt.title('Apple per-stream payment')
plt.xlabel('Per-stream payment ($)')
plt.show()

sns.displot(spotify.Price, kde=True)
plt.title('Spotify per-stream payment')
plt.xlabel('Per-stream payment ($)')
plt.show()
```

```python
# Violin plot
data = df[(df['Partner Name'] == 'Apple') | (df['Partner Name'] == 'Spotify')]
sns.violinplot(data=data, x='Price', y='Partner Name')
```

```python
# Box plot
sns.boxplot(data=data, x='Price', y='Partner Name')
```

```python
print("Median Apple Payout: ", round(np.median(apple.Price), 5))
print("Median Spotify Payout: ", round(np.median(spotify.Price), 5))
```

```python
# Test significance
import scipy.stats as stats
_, p = stats.mannwhitneyu(apple.Price.values, spotify.Price.values, alternative='two-sided')

print(p)
```

```python
top_countries = df['Delivery Country'].value_counts()[0:5].index.values

sns.boxplot(data=apple[apple['Delivery Country'].isin(top_countries)], x='Price', y='Delivery Country')
plt.show()

sns.boxplot(data=spotify[spotify['Delivery Country'].isin(top_countries)], x='Price', y='Delivery Country')
```
